```python
import pandas as pd

data = {
    'Sistem': ['A']*8 + ['B']*8 + ['C']*8,
    'Bahan': (['X']*2 + ['Y']*2 + ['Z']*2 + ['W']*2)*3,
    'Laju': [
        34.0, 32.7, 30.1, 31.8, 29.8, 26.7, 29.0, 28.9,   # Sistem A
        32.0, 33.2, 30.2, 29.8, 28.7, 28.1, 27.9, 27.7,   # Sistem B
        28.4, 29.3, 27.0, 28.9, 29.7, 27.3, 28.8, 29.1    # Sistem C
    ]
}

df = pd.DataFrame(data)
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Laju ~ C(Sistem) + C(Bahan) + C(Sistem):C(Bahan)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Data diambil dari tabel
data = {
    'C2F6_Flow_Rate': [1]*6 + [2]*6 + [3]*6,
    'Power_Supplied': [1,1,2,2,3,3]*3,
    'Kecepatan_Etsa': [
        288, 360, 488, 465, 670, 720,   # Flow rate 1
        385, 411, 482, 521, 692, 724,   # Flow rate 2
        488, 462, 595, 612, 761, 801    # Flow rate 3
    ]
}

df = pd.DataFrame(data)

# Lakukan ANOVA dua arah
model = ols('Kecepatan_Etsa ~ C(C2F6_Flow_Rate) + C(Power_Supplied) + C(C2F6_Flow_Rate):C(Power_Supplied)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
